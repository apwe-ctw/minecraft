# (Optional) If you want to save the output to a video

dont make the video too long, since the size can be VERY BIG

```python
# Open the camera
cam = cv2.VideoCapture(0)  # 0 is the default camera. If you have multiple camera, you might need different values

# Get the default frame width and height
frame_width = int(cam.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cam.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 20.0, (frame_width, frame_height))

while True:
    ret, frame = cam.read()
    # Write the frame to the output file
    out.write(frame)

    cv2.imshow('Camera', frame)
    # Press 'q' to exit the loop
    if cv2.waitKey(1) == ord('q'):
        break

# Release the capture and writer objects
cam.release()
out.release()
cv2.destroyAllWindows()
```

Change the camera setting one-by-one by trial and errors
- Brightness, Constrast, Saturation
- default is - 0.5, 0.5, 0.5
- It accepts float values from 0.0 to 1.0 (Question: what is the difference between 1 and 1.0?)

Tune them untill you think the color is good enough. There are many different scenarios (e.g. static / motion imaging , bright day / dark night setting, interference from other EM waves)

It is General very hard to tune a camera to be like real colors in all scenarios. And real color might not always be the most attractive one. E.g. LG tends to tune the cell phone camera with higher saturation, to give more refreshing feel in the image.

Adjust those parameters with your perference

```python
# Display the face Image

cv2.imshow("Face", face_img)
while True:
    if cv2.waitKey(1) == ord('q'):
        cv2.destroyAllWindows()
        break
```

```python
face_classifier = cv2.CascadeClassifier("./haarcascade_frontalface_default.xml.1")

def detect_bounding_box(vid):
    gray_image = cv2.cvtColor(vid, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray_image, 1.1, 5, minSize=(40, 40))
    for (x, y, w, h) in faces:
        cv2.rectangle(vid, (x, y), (x + w, y + h), (0, 255, 0), 4)
    return faces
cam = cv2.VideoCapture(0)

cam.set(cv2.CAP_PROP_BRIGHTNESS, 0.7) # values: 0-1
cam.set(cv2.CAP_PROP_CONTRAST, 0.5) # values: 0-1
cam.set(cv2.CAP_PROP_SATURATION, 0.65) # valuesL 0-1

while True:
    # Read the camera
    ret, frame = cam.read()
    try:
        faces = detect_bounding_box(frame)  # apply the function we created to the video frame
    except Exception as e:
        print("Error", e)

    # Display the captured frame
    cv2.imshow('Camera', frame)
    
    # Press 'q' to exit the loop
    if cv2.waitKey(1) == ord('q'):
        break  # break the while loop

cam.release()
cv2.destroyAllWindows()
```
